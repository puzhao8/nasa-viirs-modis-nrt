```python
import os, requests
from pathlib import Path
from bs4 import BeautifulSoup
from prettyprinter import pprint

def get_url_paths(url, ext='', params={}):
    response = requests.get(url, params=params)
    if response.ok:
        response_text = response.text
    else:
        return response.raise_for_status()
    soup = BeautifulSoup(response_text, 'html.parser')
    parent = [url + node.get('href') for node in soup.find_all('a') if node.get('href').endswith(ext)]
    return parent



dataPath = Path("/content/drive/MyDrive/VIIRS/VNP09GA.001")
if not os.path.exists(dataPath): os.makedirs(dataPath)

url_root = 'https://e4ftl01.cr.usgs.gov/VIIRS/VNP09GA.001/2021.07.06/'
ext = '.h5'
url_list = get_url_paths(url_root, ext)

# for url in result:
#     print(url)

save_folder =  dataPath / os.path.split(url_root)[0].split("/")[-1]
if not os.path.exists(save_folder): os.makedirs(save_folder)
os.chdir(save_folder)

hHHvVV = [f"h{HH}v{VV}" for VV in ["02", "03", "04", "05", "06"] for HH in ["08", "09", "10", "11", "12", "13"]]
to_download = [url for url in url_list if os.path.split(url)[-1].split(".")[-4] in hHHvVV]

# 
for url in to_download:
    print(url)
    !wget --user="puzhao_agb" --password "kth10044NASA!" $url

```
